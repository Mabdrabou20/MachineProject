```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import warnings
```

```python
data_old = pd.read_csv("Automobile_data.csv")
data_old

```

```python
data=data_old.drop(['engine-location' ,'engine-location','price','fuel-type','make','num-of-doors'], axis=1)
data
```

```python
data.replace('?',np.nan, inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data['normalized-losses'] = data['normalized-losses'].replace(np.NaN, data['normalized-losses'].mode()[0])
data['bore'] = data['bore'].replace(np.NaN, data['bore'].mode()[0])
data['stroke'] = data['stroke'].replace(np.NaN, data['stroke'].mode()[0])
data['horsepower'] = data['horsepower'].replace(np.NaN, data['horsepower'].mode()[0])
data['peak-rpm'] = data['peak-rpm'].replace(np.NaN, data['peak-rpm'].mode()[0])
```

```python
data.isnull().sum()
```

```python
x=data.drop(["symboling"],axis=1)
x.head()
```

```python
y=data["symboling"]
y.head()
```

```python
x.info()
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols=['normalized-losses','aspiration' ,'body-style','drive-wheels','engine-type','num-of-cylinders','fuel-system','bore','stroke','horsepower','peak-rpm'])
x=encoder.fit_transform(x)
x.head(200)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
model=DecisionTreeClassifier(max_depth=9)
```

```python
model.fit(x_train,y_train)
```

```python
tree_y_pre=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,tree_y_pre)
print("model accuracy score:", accuracy*100 ," %")
```
