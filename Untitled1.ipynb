```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data_old= pd.read_csv("Automobile_data.csv")
data_old
```

```python
data=data_old.drop(['engine-location' ,'price','fuel-type','make','num-of-doors','aspiration'], axis=1)
data
```

```python
data.replace('?', np.nan, inplace=True)
data
```

```python
data.info()
```

```python
import category_encoders as ce
encoder=ce.OrdinalEncoder(cols=['normalized-losses','body-style','drive-wheels','engine-type','num-of-cylinders','fuel-system','bore','stroke','horsepower','peak-rpm'])
data=encoder.fit_transform(data)
data
```

```python
data.isnull().sum()
```

```python
data['normalized-losses'] = data['normalized-losses'].replace(np.NaN, data['normalized-losses'].mean())
data['bore'] = data['bore'].replace(np.NaN, data['bore'].mean())
data['stroke'] = data['stroke'].replace(np.NaN, data['stroke'].mean())
data['horsepower'] = data['horsepower'].replace(np.NaN, data['horsepower'].mean())
data['peak-rpm'] = data['peak-rpm'].replace(np.NaN, data['peak-rpm'].mean())
```

```python
data.isnull().sum()
```

```python
x=data.drop(["symboling"],axis=1)
x.head()
```

```python
y=data["symboling"]
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy*100)
```
